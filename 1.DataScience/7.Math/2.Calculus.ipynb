```python
# Dung bang ASCII de check characters

def countCharactersInString(a):
    lowers = 0
    uppers = 0
    digits = 0

    # Use ASCII Table to check characters
    for i in a:
        ascii_num = ord(i)
        if 97 <= ascii_num <= 122:
            lowers += 1
        elif 65 <= ascii_num <= 90:
            uppers += 1
        elif 48 <= ascii_num <= 57:
            digits += 1
    
    # total letters
    letters = lowers + uppers

    # other characters beside letters & digits
    others = len(a) - lowers - uppers - digits

    return [letters, lowers, uppers, digits, others]
            
```
